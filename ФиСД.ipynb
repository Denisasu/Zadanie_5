### Задание 1
Переведите содержимое файла purchase_log.txt в словарь purchases вида:
{‘1840e0b9d4’: ‘Продукты’, …}


Пример работы программы при выводе первых двух элементов словаря purchases:
1840e0b9d4 ‘Продукты‘
4e4f90fcfb ‘Электроника‘

```python
# === ЗАДАНИЕ 1 ===
# Переведите содержимое файла purchase_log.txt в словарь purchases

import json

# Создаём пустой словарь
purchases = {}

# Открываем файл и читаем построчно
with open('purchase_log.txt', 'r', encoding='utf-8') as file:
    for line in file:
        # Убираем лишние пробелы и переносы строк
        line = line.strip()
        
        # Пропускаем пустые строки
        if not line:
            continue
        
        # Преобразуем строку JSON в словарь Python
        data = json.loads(line)
        
        # Добавляем в наш словарь: ключ - user_id, значение - category
        user_id = data['user_id']
        category = data['category']
        purchases[user_id] = category

# Выводим первые два элемента словаря (как в примере)
print("Первые две записи словаря purchases:")
count = 0
for user_id, category in purchases.items():
    print(f"{user_id} '{category}'")
    count += 1
    if count == 2:
        break

# Дополнительно: покажем сколько всего записей
print(f"\nВсего записей в словаре: {len(purchases)}")
```

### Задание 2 (необязательное)

Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки. Если покупка была, сам файл visit_log.csv изменять не надо.

Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.
Учтите условия для данных:

* содержимое purchase_log.txt помещается в оперативную память компьютера;
* содержимое visit_log.csv — нет, используйте только построчную обработку этого файла.
Примеры работы программы при выводе первых трёх строк файла funnel.csv:
user_id,source,category
1840e0b9d4,other,Продукты
4e4f90fcfb,context,Электроника

```python
# === ЗАДАНИЕ 2 ===
# Создаём funnel.csv из visit_log.csv с добавлением категории покупки

# Функция уже создана в ячейке 1, поэтому просто используем purchases

# Открываем оба файла одновременно
with open('visit_log.csv', 'r', encoding='utf-8') as visit_file, \
     open('funnel.csv', 'w', encoding='utf-8') as funnel_file:
    
    # Читаем заголовок из visit_log.csv
    header = visit_file.readline().strip()
    
    # Проверяем, что это действительно заголовок с user_id
    if header.startswith('user_id'):
        # Пишем новый заголовок в funnel.csv
        funnel_file.write('user_id,source,category\n')
    else:
        # Если первая строка не заголовок, обрабатываем её как данные
        funnel_file.write('user_id,source,category\n')
        # Разделяем строку по запятым
        parts = header.split(',')
        if len(parts) >= 2:
            user_id = parts[0]
            source = parts[1]
            # Если была покупка - записываем
            if user_id in purchases:
                category = purchases[user_id]
                funnel_file.write(f'{user_id},{source},{category}\n')
    
    # Обрабатываем остальные строки из visit_log.csv
    for line in visit_file:
        line = line.strip()
        if line:  # пропускаем пустые строки
            # Разделяем строку по запятым
            parts = line.split(',')
            if len(parts) >= 2:
                user_id = parts[0]
                source = parts[1]
                
                # Проверяем, есть ли покупка у этого пользователя
                if user_id in purchases:
                    category = purchases[user_id]
                    # Записываем в funnel.csv
                    funnel_file.write(f'{user_id},{source},{category}\n')

print("Файл funnel.csv успешно создан!")
```

```python
# === ПРОВЕРКА ===
# Выводим первые 3 строки созданного файла funnel.csv

print("Первые 3 строки файла funnel.csv:")
print("-" * 40)

with open('funnel.csv', 'r', encoding='utf-8') as f:
    lines_printed = 0
    for line in f:
        print(line.strip())
        lines_printed += 1
        if lines_printed >= 3:
            break

print("-" * 40)

# Посчитаем сколько всего строк в файле
with open('funnel.csv', 'r', encoding='utf-8') as f:
    total_lines = sum(1 for line in f)
    print(f"Всего строк в funnel.csv: {total_lines}")
```
